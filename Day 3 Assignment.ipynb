```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fmri=sns.load_dataset("fmri")
```

```python
fmri.info()
```

```python
fmri.describe()
```

```python
sns.relplot(x="timepoint",y="signal",hue="timepoint",data=fmri)
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event")
plt.title("LinePlot of Fmri dataset")
```

```python
sns.scatterplot(x="timepoint",y="signal",data=fmri,hue="region")
plt.title("ScatterPlot of Fmri dataset")
```

```python
sns.catplot(x="timepoint",y="signal",data=fmri,hue="region",dodge=False)
plt.title("catPlot of Fmri dataset",color="red")
```

```python
sns.boxplot(x="timepoint",y="signal",data=fmri,hue="event")
plt.title("Box plot for fmri dataset",color="g")
```
