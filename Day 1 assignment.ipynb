<h1>QUESTION 1</h1>

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,10)
y=x*x
plt.title("Question 1 using Line plot",color="green")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,y,'y--')
plt.show()
```

<h1>Question 2</h1>

```python
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
plt.title("Line plot",color="red")
plt.xlabel('days of week',color="red")
plt.ylabel('sales of company',color="red")
plt.plot(days,sales_1,"go-")
plt.plot(days,sales_2,"b*-")
```
